```python
from project.thermal import conventional_heating_final, stat_heating_consumption, HDD, conventional_energy_3uses
from project.utils import get_pandas
from project.input.param import generic_input
import pandas as pd
import os
```

```python

buildings = get_pandas('project/input/stock/building_stock_profeel.csv', lambda x: pd.read_csv(x, index_col=[0, 1]).squeeze())
ratio_surface = get_pandas('project/input/stock/ratio_surface_profeel.csv', lambda x: pd.read_csv(x, index_col=[0, 1]).squeeze())
ratio_surface = ratio_surface.loc[:,'Ratio_surface_Wall':]
ratio_surface = ratio_surface.rename(columns={'Ratio_surface_Floor': "Floor", 'Ratio_surface_Wall': "Wall", 'Ratio_surface_Roof': "Roof", 'Ratio_surface_Window': "Windows"})
buildings
```

```python
efficiency = {'Electricity-Performance boiler': 0.95,
           'Natural gas-Standard boiler': 0.6,
          'Natural gas-Performance boiler': 0.76,
              'Wood fuel-Standard boiler': 0.75
           }
efficiency = buildings['Heating system'].replace(efficiency).set_axis(buildings.index)
buildings = pd.concat((buildings, efficiency.rename('Efficiency')), axis=1)
buildings
```

```python
conventional_heating_final(buildings['Wall'], buildings['Floor'], buildings['Roof'], buildings['Windows'], ratio_surface, buildings['Efficiency'])
```

```python
stat_heating_consumption(buildings['Wall'], buildings['Floor'], buildings['Roof'], buildings['Windows'], buildings['Efficiency'], ratio_surface, HDD)
```

```python
buildings = buildings.set_index('Heating system', append=True)
buildings
```

```python
epc, _ = conventional_energy_3uses(buildings['Wall'], buildings['Floor'], buildings['Roof'], buildings['Windows'], ratio_surface, buildings['Efficiency'], buildings.index)
```

```python
buildings = pd.concat((buildings.loc[:, 'Wall':'Stock buildings'], epc.rename('EPC')), axis=1)
```

```python
buildings
```

```python
path_output = os.path.join(os.getcwd(), 'preprocessing/building_stock/output')
buildings.to_csv(os.path.join(path_output, 'building_stock_profeel_epc.csv'))
```

```python
from project.thermal import conventional_heating_final, heating_consumption, HDD_EQ
from project.utils import get_pandas
from project.input.param import generic_input
import pandas as pd
```

```python
buildings = get_pandas('project/input/building_stock_profeel.csv', lambda x: pd.read_csv(x, index_col=[0, 1]).squeeze())
ratio_surface = buildings.loc[:,'Ratio_surface_Wall':]
ratio_surface = ratio_surface.rename(columns={'Ratio_surface_Floor': "Floor", 'Ratio_surface_Wall': "Wall", 'Ratio_surface_Roof': "Roof", 'Ratio_surface_Window': "Windows"})
buildings
```

```python
buildings.loc[buildings.loc[:,'Heating energy'] == 'Natural gas', 'Efficiency'] = 0.75
buildings.loc[buildings.loc[:,'Heating system'] == 'Performance boiler', 'Efficiency'] = 0.85
buildings.loc[buildings.loc[:,'Heating energy'] == 'Wood fuel', 'Efficiency'] = 0.75
buildings.loc[buildings.loc[:,'Heating energy'] == 'Electricity', 'Efficiency'] = 0.95
buildings = buildings[buildings.loc[:,'Heating energy'] != 'RCU'] #RCU archetypes are deleted
buildings = buildings.loc[:,['Wall', 'Floor', 'Roof', 'Windows', 'Efficiency']]
buildings
```

```python
conventional_heating_final(buildings['Wall'], buildings['Floor'], buildings['Roof'], buildings['Windows'], ratio_surface, buildings['Efficiency'])
```

```python
from project.thermal import model_heating_consumption, certificate, DHH, CONVERSION
from project.input.param import param, generic_input
from project.utils import reindex_mi
```

```python
ratio_surface = param['ratio_surface']
ratio_surface
```

# Transforming data

```python
data['Losses mur'] = data['R_mur']**-1 * data['S_mur'] / data['SHAB']
data['Losses plancher'] = data['R_plancher']**-1 * data['S_plancher'] / data['SHAB']
data['Losses plafond'] = data['R_plafond']**-1 * data['S_plafond'] / data['SHAB']
data['Losses baie'] = data['R_baie']**-1 * data['S_baie'] / data['SHAB']

data = data.replace(float('nan'), 0)

data['Losses'] = (data['Losses mur'] + data['Losses plancher'] + data['Losses plafond'] + data['Losses baie']) / data['Efficiency'] * DHH / 1000
```

```python
data['Consumption'] = model_heating_consumption(data['Losses'])
data['Primary consumption no elec'] = data['Consumption']
data['Primary consumption elec'] = data['Consumption'] * CONVERSION
data['Performance no elec'] = certificate(data['Primary consumption no elec'])
data['Performance elec'] = certificate(data['Primary consumption elec'])
```

### Calculate transmittance equivalent using building archetypes compacity

```python
data['Umur_equivalent'] = data['Losses mur'] / ratio_surface.loc[:, 'Wall']
data['U_wall'] = data['Umur_equivalent'].round(2)

data['Ufloor_equivalent'] = data['Losses plancher'] / ratio_surface.loc[:, 'Floor']
data['U_floor'] = data['Ufloor_equivalent'].round(2)

data['Uroof_equivalent'] = data['Losses plafond'] / ratio_surface.loc[:, 'Roof']
data['U_roof'] = data['Uroof_equivalent'].round(2)

data['Uwindows_equivalent'] = data['Losses baie'] / ratio_surface.loc[:, 'Windows']
data['U_windows'] = data['Uwindows_equivalent'].round(2)
```

### Setting-up transmittance max

```python

performance_max = generic_input['performance_insulation']
print(performance_max)

for i, u in performance_max.items():
    if i != 'Efficiency':
        name = 'U_{}'.format(i.lower())
        data.loc[data[name] < u, name] = u

```

```python
efficiency_approx = {0.958636503: 0.95,
                     0.773768981: 0.75,
                     0.831176918: 0.85,
                     3.66020518: 3,
                     0.8: 0.75,
                     0.688787035: 0.75,
                     2.280824849: 3,
                     1.672125565: 3,
                     0.884600857: 0.85,
                     0.678371065: 0.75,
                     2.807167087: 3}

data['Efficiency_approx'] = data['Efficiency'].replace(efficiency_approx)
```

```python
data.loc[:, 'Electricity'] = 0
data.loc[data['Efficiency'] >= 0.95, 'Electricity'] = 1
```

```python
ratio_surf = reindex_mi(ratio_surface, data.index)

data['U_roof'] = data['U_roof'].fillna(0)
data['U_floor'] = data['U_floor'].fillna(0)
data['Losses approx'] = (data['U_wall'] * ratio_surf.loc[:, 'Wall'] + data['U_floor'] * ratio_surf.loc[:, 'Floor'] + data['U_roof'] * ratio_surf.loc[:, 'Roof'] + data['U_windows'] * ratio_surf.loc[:, 'Windows']) / data['Efficiency_approx'] * DHH / 1000
data['Consumption approx'] = model_heating_consumption(data['Losses approx'])
data['Consumption approx PE'] = data['Consumption approx']
data.loc[data['Electricity'] == 1, 'Consumption approx PE'] = data['Consumption approx'] * CONVERSION
data['Performance'] = certificate(data['Consumption approx PE'])
```

```python
vars = ['U_wall', 'U_floor', 'U_roof', 'U_windows', 'Losses approx', 'Efficiency_approx', 'Consumption approx', 'PONDMEN']
df = data.set_index(['Performance', 'Electricity'], append=True)[vars].reorder_levels(['Housing type', 'Performance', 'Electricity', None]).sort_index()
df.index.names = ['Housing type', 'Performance', 'Electricity', 'Number']
df = df.rename(columns={'Efficiency_approx': 'Efficiency'})
df
```

## Best performance buildings archetypes

```python
performance_max['Efficiency'] = 0.85
data_best = pd.Series(performance_max).set_axis(['U_wall', 'U_floor',  'U_roof', 'U_windows', 'Efficiency'], axis=0)
data_best = pd.concat([data_best] * 2, axis=1, keys=['Single-family', 'Multi-family'], names='Housing type').T

data_best['Losses approx'] = (data_best['U_wall'] * ratio_surface.loc[:, 'Wall'] + data_best['U_floor'] * ratio_surface.loc[:, 'Floor'] + data_best['U_roof'] * ratio_surface.loc[:, 'Roof'] + data_best['U_windows'] * ratio_surface.loc[:, 'Windows']) / data_best['Efficiency'] * DHH / 1000
data_best['Consumption approx'] = model_heating_consumption(data_best['Losses approx'])
data_best['Performance'] = certificate(data_best['Consumption approx'])
data_best_noelec = data_best.set_index('Performance', append=True)

performance_max['Efficiency'] = 3
data_best = pd.Series(performance_max).set_axis(['U_wall', 'U_floor', 'U_roof', 'U_windows', 'Efficiency'], axis=0)
data_best = pd.concat([data_best] * 2, axis=1, keys=['Single-family', 'Multi-family'], names='Housing type').T

data_best['Losses approx'] = (data_best['U_wall'] * ratio_surface.loc[:, 'Wall'] + data_best['U_floor'] * ratio_surface.loc[:, 'Floor'] + data_best['U_roof'] * ratio_surface.loc[:, 'Roof'] + data_best['U_windows'] * ratio_surface.loc[:, 'Windows']) / data_best['Efficiency'] * DHH / 1000
data_best['Consumption approx'] = model_heating_consumption(data_best['Losses approx'])
data_best['Performance'] = certificate(data_best['Consumption approx'] * CONVERSION)
data_best_elec = data_best.set_index('Performance', append=True)

data_best = pd.concat((data_best_noelec, data_best_elec), keys=[0, 1], names=['Electricity'])
data_best = pd.concat([data_best], keys=[0], names=['Number']).reorder_levels(df.index.names)
data_best['PONDMEN'] = 1

data_best.loc['Multi-family', 'U_floor'] = 0
data_best.loc['Multi-family', 'U_roof'] = 0

data_best
```

```python
data_worst = pd.Series({'U_wall': df['U_wall'].max(), 'U_floor': 0, 'U_roof': 0, 'U_windows':df['U_windows'].max(), 'Efficiency': 0.85}).to_frame().T
data_worst.index = pd.Index(['Multi-family'], name='Housing type')

data_worst['Losses approx'] = (data_worst['U_wall'] * ratio_surface.loc[:, 'Wall'] + data_worst['U_floor'] * ratio_surface.loc[:, 'Floor'] + data_worst['U_roof'] * ratio_surface.loc[:, 'Roof'] + data_worst['U_windows'] * ratio_surface.loc[:, 'Windows']) / data_worst['Efficiency'] * DHH / 1000
data_worst['Consumption approx'] = model_heating_consumption(data_worst['Losses approx'])
data_worst['Performance'] = certificate(data_worst['Consumption approx'])
data_worst = data_worst.set_index('Performance', append=True)
data_worst =  pd.concat([data_worst], keys=[0], names=['Number'])
data_worst =  pd.concat([data_worst], keys=[0], names=['Electricity'])
data_worst = data_worst.reorder_levels(df.index.names)
data_worst['PONDMEN'] = 1

data_worst
```

```python
df = pd.concat((data_best, data_worst, df), axis=0)
df
```

### Formatting data

```python
result = pd.DataFrame()
for n, g in df.groupby(['Electricity', 'Performance', 'Housing type']):
    g['Proba'] = g['PONDMEN'] / g['PONDMEN'].sum()
    temp = g[['U_wall', 'U_floor', 'U_roof', 'U_windows', 'Efficiency', 'Proba']]
    if n[0] == 0:
        temp = pd.concat([temp] * 3, keys=['Natural gas', 'Oil fuel', 'Wood fuel'], names=['Heating energy'], axis=0)
    else:
        temp = pd.concat([temp], keys=['Electricity'], names=['Heating energy'], axis=0)

    result = pd.concat((result, temp), axis=0)
```

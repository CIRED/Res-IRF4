## Plot Hidden Cost Sensitivity Results

**Purpose**
Plots an illustrative hidden-cost distribution using calibrated parameters (hidden cost level, scale, and dispersion assumptions).

**Primary inputs**
- Hard-coded calibration parameters in this notebook (`hidden_cost`, `coefficient_cost`, `scale_model`)

**Primary outputs**
- In-notebook distribution plot


## Execution Notes

- Run all cells top-to-bottom from a clean kernel.
- Update the three calibration parameters in the code cell to test alternative assumptions.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import logistic

hidden_cost = 49

coefficient_cost = 0.0964

scale_model = 0.738956596364766

# Number of samples
num_samples = 10000

# Generate samples from Gumbel distribution
gumbel_samples = logistic.rvs(size=num_samples)


# Plot histograms
plt.figure(figsize=(10, 5))

plt.hist(gumbel_samples / coefficient_cost * scale_model + hidden_cost, bins=30, density=True, alpha=0.7, color='b')
plt.title('Distribution of hidden cost for representative insulation project (Thousands of euro)')
plt.xlabel('Value')
plt.ylabel('Density')

plt.tight_layout()
plt.show()
```
